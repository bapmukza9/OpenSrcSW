```python
# 실습 1: 생일 기온 시각화
import matplotlib.pyplot as plt
import csv
f = open('seoul_decade.csv')
data = csv.reader(f)
next(data)
high_result = []
low_result = []
for row in data:
    if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '03':
        high_result.append(float(row[-1]))
        low_result.append(float(row[-2]))

year = list(range(2011,2021))

plt.rc('font', family='Malgun Gothic')
plt.title('10년간 6월 3일(생일) 최고·최저기온')
plt.plot(year, high_result, 'pink', label='최고기온')
plt.plot(year, low_result, 'skyblue', label='최저기온')
plt.legend()
plt.show()
```

```python
# 실습 2: 최고 기온 시각화
import matplotlib.pyplot as plt
import csv
f = open('seoul_decade.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data:
    month=row[0].split('-')[1]
    if month == '01':
        jan.append(float(row[-1]))
    elif month == '08':
        aug.append(float(row[-1]))
        
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus = False)
plt.title('10년간 서울 지역 1월과 8월 최고기온 히스토그램')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.legend()
plt.show()
```
